```
Given a string of digits, return the longest substring with alternating odd/even or even/odd digits. If two or more substrings have the same length, return the substring that occurs first.

Examples
longest_substring("225424272163254474441338664823") ➞ "272163254"
# substrings = 254, 272163254, 474, 41, 38, 23

longest_substring("594127169973391692147228678476") ➞ "16921472"
# substrings = 94127, 169, 16921472, 678, 476

longest_substring("721449827599186159274227324466") ➞ "7214"
# substrings = 7214, 498, 27, 18, 61, 9274, 27, 32
# 7214 and 9274 have same length, but 7214 occurs first.
```

````python
text = "225424272163254474441338664823"
liste = []
if int(text[0]) % 2 == 0:
    odd = True
else:
    odd = False
if int(text[-1]) % 2 == 0:
    text += '0'
else:
    text += '1'
yeni = text[0]
for i in text[1:]:
    if odd:
        if int(i) % 2 != 0:
            yeni += i
            odd = False
        else:
            if len(yeni) > 1:
                liste.append(yeni)
            yeni = str(i)
    else:
        if int(i) % 2 == 0:
            yeni += i
            odd = True
        else:
            if len(yeni) > 1:
                liste.append(yeni)
            yeni = str(i)
if liste != []:
    result = liste[0]
    for i in liste:
        if len(i) > len(result):
            result = i
else:
    result = liste
print(liste)
print(result)
````

````python
def alternating(text):
    liste = []
    if int(text[0]) % 2 == 0:
        odd = True
    else:
        odd = False
    if int(text[-1]) % 2 == 0:
        text += '0'
    else:
        text += '1'
    yeni = text[0]
    for i in text[1:]:
        if odd:
            if int(i) % 2 != 0:
                yeni += i
                odd = False
            else:
                if len(yeni) > 1:
                    liste.append(yeni)
                yeni = str(i)
        else:
            if int(i) % 2 == 0:
                yeni += i
                odd = True
            else:
                if len(yeni) > 1:
                    liste.append(yeni)
                yeni = str(i)
    if liste != []:
        result = liste[0]
        for i in liste:
            if len(i) > len(result):
                result = i
    else:
        result = liste
    return liste, f'The longest alternating substring is: {result}'

alternating("225424272163254474441338664823")
````
